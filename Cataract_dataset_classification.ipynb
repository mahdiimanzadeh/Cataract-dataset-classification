```python
train_dir = '/content/dataset/train'
```

```python
# Get a list of all subfolders in the train directory
import os
subfolders_train = [f for f in os.listdir(train_dir) if os.path.isdir(os.path.join(train_dir, f))]
subfolders_train
```

```python
total_image_count_train = 0  # Initialize total image count

for folder in subfolders_train:
    folder_path = os.path.join(train_dir, folder)
    image_count = 0

    for file in os.listdir(folder_path):
        if file.lower().endswith(('.png', '.jpg', '.jpeg')):
            image_count += 1

    print(f"Folder: {folder}, Image Count: {image_count}")
    total_image_count_train += image_count  # Add subfolder's count to total

print(f"\nTotal Images: {total_image_count_train}")
```

```python
import matplotlib.pyplot as plt
import cv2
import random
from glob import glob

# پیدا کردن تصاویر یک کلاس (مثلاً Cornea)
image_files = glob('/content/dataset/train/Cornea Lens Injector/*')

# نمایش یکی از تصاویر تصادفی
if image_files:
    sample_image = random.choice(image_files)
    img = cv2.imread(sample_image)
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.title(f"Sample Image from Cornea Class")
    plt.axis("off")
    plt.show()
else:
    print("No images found in the specified class.")
```
